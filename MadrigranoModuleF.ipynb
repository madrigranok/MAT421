```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
h = .1
x = np.arange(0, 2*np.pi, h)
y = np.cos(x)
fd = np.diff(y)/h
xdiff = x[:-1:]
exact = -np.sin(xdiff)
error = max(abs(exact-fd))
plt.plot(xdiff, fd, '--', label = "Finite Approx")
plt.plot(xdiff, exact, label = "Exact Solution")
plt.legend()
plt.show()
print(error)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
h = 1
it = 20
step = []
error = []
for i in range(it) :
  h /=2
  step.append(h)
  x = np.arange(0, 2*np.pi, h)
  y = np.cos(x)
  fd = np.diff(y)/h
  xdiff = x[:-1]
  exact = -np.sin(xdiff)
  error.append(max(abs(exact-fd)))
plt.figure(figsize = (12, 8))
plt.loglog(step, error, 'v')
plt.show()
```

Noise

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
x = np.arange(0, 2*np.pi, 0.01)
omega = 100
epsilon = 0.01
y = np.cos(x)
ynoise = y + epsilon*np.sin(omega*x)
plt.plot(x, ynoise, 'r-', label = "cos(x) with noise")
plt.plot(x, y, 'b-', label = "cos(x)")
plt.legend()
plt.show()
```

Derivative

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
x = np.arange(0, 2*np.pi, 0.01)
omega = 100
epsilon = 0.01
y = -np.sin(x)
ynoise = y + epsilon*omega*np.cos(omega*x)
plt.plot(x, ynoise, 'r-', label = "sin(x) with noise")
plt.plot(x, y, 'b-', label = "sin(x)")
plt.legend()
plt.show()
```
