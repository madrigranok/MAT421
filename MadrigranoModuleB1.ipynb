```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster') #I dont particularly understand what this means but I'm using it since the code for the notes had it

#I want to interpolate the function fx = x^3 -2x -2 for 5 data points since this is the one I've been learning interpolation with in num analysis 2

x = [-2, -1, 0, 1, 2]
y = [-2, 3, 2, 1, 6]

f = interp1d(x,y)
yhat = f(1.5)
print(yhat)

plt.plot(x, y, "-ob")
plt.plot(1.5, yhat, "ro")
```

Now moving onto the cubic interpolation

```python
from IPython.core.display import clear_output
clear_output
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [-2, -1, 0, 1, 2]
y = [-2, 3, 2, 1, 6]
f = CubicSpline(x, y, bc_type='natural')
xnew = np.linspace(-2, 2, 100)
ynew = f(xnew)

plt.plot(xnew, ynew, "b")
plt.plot(x, y, "ro")
```
