```python
#continuity and differentiation
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')  #from berkeley's stuff

h = 0.2
x = np.arange(0, 2*np.pi, h)
y = np.sin(x)

fdif = np.diff(y)/h
xdif = x[:-1:]
esol = np.cos(xdif)

plt.plot(xdif, fdif, '--', label = "Finite Difference Approx")
plt.plot(xdif, esol, label = "Exact Solution")
plt.legend()
plt.show()
```

```python
#taylor series top tier analysis
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ["1st order", "2nd order"]

for n, label in zip(range(2), labels):
  y = y + ((-1)**n * (x)**(2*n+1))/ np.math.factorial(2*n+1)
  plt.plot(x,y, label = label)

plt.plot(x, np.sin(x), "k", label = "Analytic")
plt.grid()
plt.title("Taylor Series Approx")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(0,8)
y = x**2

x1 = np.linspace(0,6)
y1 = x*4.5

plt.plot(x, y, "g", x1, y1, "r")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.arange(0, np.pi, 0.01)
Fe = -np.cos(x)
Fa = cumtrapz(np.sin(x), x)

plt.plot(x, Fe)
plt.plot(x[1::], Fa)
plt.grid()
plt.title('$F(x) = \int_0^{x} sin(y) dy$')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["Exact with Offset", "Approx"])
plt.show()
```
